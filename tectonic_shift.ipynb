```python
import numpy as np
from matplotlib import pyplot as plt

csv_arr = np.genfromtxt('/home/zeelb/Downloads/volcano_data.csv', delimiter=',')

data = csv_arr[:, 2:4]
x = data[:, 1]
y = data[:, 0]

m = sum((x - np.mean(x))*(y - np.mean(y)) / sum((x - np.mean(x))**2))
b = np.mean(y) - (m * np.mean(x))
residual_standard_error = int(np.sqrt(np.sum((y - (m*x) - b)**2)/ (np.shape(x)[0] - 2)))
print("Residual Standard Error : " + str(residual_standard_error) + " Kilometers")
line_x = np.linspace(0, 70, 100)
line_y = (m*line_x) + b

x = data[:, 1]
y = data[:, 0]

fig, ax = plt.subplots()
ax.scatter(x, y)
plt.plot(line_x, line_y, '-r')
plt.xlabel('Years in millions', color='#1C2833')
plt.ylabel('Kilometers', color='#1C2833')
plt.grid()
plt.show()
```
